# **New feature (Level 5 visitors/Level 1 visitors)**

Below is a table showing top 3 states based on the created feature

```python
df1['Lv5_Lv1_Ratio'] = df1['Lv5_Visitors']/df1['Lv1_Visitors']

# Group by Year, Segment, and State and compute mean of new feature
df_grouped = df1.groupby(['Segment','Year', 'Region'])['Lv5_Lv1_Ratio'].mean().reset_index()

# Sort by new feature in descending order
df_sorted = df_grouped.sort_values([ 'Segment','Year', 'Lv5_Lv1_Ratio'], ascending=[True, True, False])

# Group by Year and Segment and find top 3 states for each group based on new feature
df_top3 = df_sorted.groupby([ 'Segment','Year']).apply(lambda x: x.nlargest(3, 'Lv5_Lv1_Ratio'))
df_top3=df_top3.drop(['Segment','Year'],axis=1)
```

```python
print(df_top3)
```

# **Another Feature Creation**

Based on this new sum metric the top 3 states of different Segments and Years


```python
df1["Total Visitors"] = df1["Lv1_Visitors"] + df1["Lv2_Visitors"] + df1["Lv3_Visitors"] + df1["Lv4_Visitors"] + df1["Lv5_Visitors"]

# calculate new metric (Level 5 visitors/total visitors)
df1["New Metric"] = df1["Lv5_Visitors"] / df1["Total Visitors"]

# find top 3 states based on new metric for each year and segment
for year in df1["Year"].unique():
    for segment in df1["Segment"].unique():
        temp_df = df1[(df1["Year"]==year) & (df1["Segment"]==segment)]
        temp_df = temp_df.groupby(["Region"]).agg({"New Metric": "mean"}).reset_index()
        temp_df = temp_df.sort_values("New Metric", ascending=False).head(3)
        print(f"Top 3 states for {segment} in {year}:")
        print(temp_df)
        print("\n")
```

The reason for this difference could be that the new metric takes into account the total number of visitors in each segment and region, while the Level 5 visitors/Level 1 visitors metric only considers the number of visitors in the two levels. This means that regions with high overall visitor counts might not necessarily have high Level 5 visitor counts, and vice versa.

Additionally, the new metric may give more weight to regions with higher overall visitor counts, which could impact the ranking of states.

Therefore, it is possible that the states in the top 3 for the new metric may be different from the top 3 states based on the Level 5 visitors/Level 1 visitors metric. However, without analyzing the data and comparing the results, it is difficult to say for certain.

```python
df3=pdf.drop(['Clients',	'Customers',	'Aurangabad',	'Dehradun',	'Faridabad',	'India',	'Indore',	"Uddepy",	'Ujjain',	'Actuals'],axis=1)
df3=df3.groupby(['Year','Month']).mean().reset_index()
```

```python
import statsmodels.api as sm
df3['Date'] = pd.to_datetime(df3['Year'].astype(str) + '-' + df3['Month'].astype(str))
df3 = df3.set_index('Date')
model = sm.tsa.ARIMA(df3['Lv5_Visitors'], order=(1,0,51))
results = model.fit()
forecast_values = results.forecast(steps=6)
forecast_index = pd.date_range(df3.index[-1], periods=6, freq='M')

# Set the forecast index to dates
forecast_values.index = forecast_index
```

```python
forecast_values
```

**Plot a line graph of the level 5 actual numbers from 2020-2022 & in the same graph, there should be the predicted numbers for 2023. The x-axis should be the timeline from 2020 Jan to 2023 Jun and the y-axis should be the value of the level 5 column and predicted values**

```python
plt.figure(figsize=(20,6))
plt.plot(df3.index, df3['Lv5_Visitors'], label='Actual')
plt.plot(forecast_index, forecast_values, label='Predicted')
plt.title('Level 5 Visitors Forecast for Next 6 Months')
plt.xlabel('Date')
plt.ylabel('Level 5 Visitors')
plt.legend()

# Set tick frequency to monthly
plt.xticks(pd.date_range(df3.index[0], forecast_index[-1], freq='MS'), rotation=70)

plt.show()
```

```python
actual = df3.loc[(df3.index.year >= 2020) & (df3.index.year <= 2022), 'Lv5_Visitors']
actual.shape
```

```python
predictions = results.predict(start='2020-01-01', end='2022-12-01')
predictions
```

# MAPE and RMSE 

```python
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error


mape = mean_absolute_percentage_error(actual, predictions)
rmse = mean_squared_error(actual, predictions, squared=False)

print('MAPE: {:.2f}%'.format(mape*100))
print('RMSE: {:.2f}'.format(rmse))
```

# **Part 4: A/B testing**

```python
# Calculate the conversion rate for each variant
control_group = df1
experimental_group = predictions

# Define the metric you want to analyze
metric = 'Total Visitors'


# Perform A/B testing using t-test
t_stat, p_value, Degree_Freedom = sm.stats.ttest_ind(control_group[metric], experimental_group, alternative='two-sided')

p_value=round(p_value,2)
print("A/B Testing Results:")
print("T-statistic: {:.4f}".format(t_stat))
print("p-value: {:.2f}".format(p_value))

alpha = 0.01
if p_value < alpha:
    print("The difference in means is statistically significant.")
else:
    print("The difference in means is not statistically significant.")
```
